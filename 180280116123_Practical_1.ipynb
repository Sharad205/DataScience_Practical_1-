# Descriptive stats describing observations
1] Mean    
2] Median   
3] Sum    
4] Max, Min  

```python
#counting students
st_count = stud_data.count().Name
print(f'Total Numer Of Students: {st_count}')
```

```python
sub = ["PEM", "OS", "COA", "DMATHS", "OOPJ"]
#getting subjects data-frame
sub_df = stud_data.loc[:, sub]
print(sub_df)

#finding mean,median,max and min using agg()
sub_df.agg([np.mean, np.median, np.min, np.max])
```

# Descriptive stats describing variable spread
1) Variance  
2) Standard deviation  
3) Counts  
4) Quartiles  

```python
#counting number of male and female students
gd_counts = stud_data.Gender.value_counts()
gd_counts.index=["Total No. of Females", "Total No. of Males"]
print(gd_counts)
```

```python
#getting gender by proportion
norm_gd_count = stud_data.Gender.value_counts(normalize=True)
norm_gd_count.index=["Females(%)", "Males(%)"]
print(norm_gd_count*100)
```

```python
#variance in marks of subject
stud_data.var()["PEM":]
```

```python
#standard deviation in marks of subject
stud_data.std()["PEM":]
```

```python
def iqr(column):
    """
    This function finds the inter quartile range.
        75th percentile - 25th percentile
    
    Parameters
    -----------
        column: column whose iqr is to be known
        
    Returns
    ------------
    the inter quartile range of specified column
    """
    return column.quantile(.75) - column.quantile(.25)
```

```python
#finding iqr of every subject marks
sub_df.agg(iqr)
```
